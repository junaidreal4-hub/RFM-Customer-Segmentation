# 05 · Tableau Export
Flatten the real scored RFM table into a single tidy CSV for Tableau/Power BI.

```python
import pandas as pd

rfm = pd.read_csv("../Data/Processed/rfm_scored_segments.csv")
cols = ["CustomerID","Recency","Frequency","Monetary",
        "R_score","F_score","M_score","FM","RFM_score","Segment"]
tableau = rfm[cols].copy()
tableau.to_csv("../Data/Processed/tableau_data.csv", index=False)
print("Saved ../Data/Processed/tableau_data.csv", tableau.shape)
tableau.head()
```
